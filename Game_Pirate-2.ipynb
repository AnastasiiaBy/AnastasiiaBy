```python
###  We are starting with time module import as during the whole game we will need
#  to prolong time for reading long text##


import time

###Defining the first function which is our start() point of the game

#defining a function  
def start(): #defining a function
    print( 80 * '*') #decoration print
 

    print('''Hi, Jack Sparrow! 
Welcome to  "Pirates of the Carribean" Quizz!''')###greetings
    time.sleep( 2.4 ) #wait until user finish reading and move to the next part 

    print ( """
As usuall you are in trouble.
The British Galleon has almost reached your ship! 
Do you want to survive? If yes - I think it is time to move.
""" )
    
    

    print( 80 * '*') #decoration 
    input(prompt = "Press Enter to start!") #first output from the user
    start_intro() #from this point we are moving to the next function
    
###  Python asks user to answer some questions in order to create a 'mad lib' story
#This stage has been created just for fun. There is no score or connection with another
#part of the game

def start_intro ():###definition of the function
    print ( 80 * '*' ) 
    print ('But first ... Tell me some interesting facts about yourself')
    list = [] #empty list in order to fill it by user's answers
    list.append ( input ( prompt ='<' "What is your favourite adjective? Please type and press enter > "))###press enter
### User can type integers instead of strings or have some grammar mistakes. 
# It will not influence the game as this part is a simple introduction


    list.append( input ( prompt = '<' "What is your favourite female name? Please type and press enter > "))
    list.append( input ( prompt = '<' "What is your favourite animal? Please type and press enter > "))
    list.append( input ( prompt = '<' "What is your favourite colour? Please type and press enter > "))
    list.append( input ( prompt = '<' "What is your favourite fruit? Please type and press enter > "))
    print ( 80 * '*' )
    
### We are using the dinamic print (values from our list which we created before the 
#questions) 
    print ( f'''Amazing, now you are leading your ship which is called '{list[0].upper()} {list[1].upper()}'. 
Your boatswain's name is {list[2].upper()}. 
You are going to "{list[3].upper()} island" in order to find gold {list[4].upper()}!!!''')###filling the statement 
   
    input("Press Enter to continue!") ###I used lists however we can also print input
    
    first_move()
###First move fubction is about conditional statement where user should choose some options 


def first_move(): # next function
    print ( 80 * '*')
    print ('''Jack, in order to escape from the Galleon you need help from your old friends! 
Who can help you?''')
    time.sleep(1.4)
    
# Four possible options (user can choose the number or the name, otherwise Python
#will mentions that the input is wrong)   
    print("""
    1.I will ask Calipso
    2.I will call Kracken
    3.I will ask Blackbeard
    4.I will fight with the Galleon""") 
    
    print ( 80 * '*')       
    choice = input ( prompt= '<').upper()#imput of user - we changes it in order to 
#compare with our answers
          
    if '1' in choice or 'CALIPSO' in choice:
        print('''No, no, no, Jackie! 
Never ever ask women for help! 
She was in a bad mood and destroyed your boat!
        ''')
        fail() # failed option 
        
    elif '2' in choice or "KRACKEN" in choice:
        print('''Are you kidding? 
Jackie, you owned some souls to old David Johnes and you want to ask his pet to help you? 
Bye, Jackie!
        ''')
        fail() # failed option
        
    elif '4' in choice or "FIGHT" in choice or 'GALLEON' in choice:
        print('''You are alone on this boat with your boatswain! Heroistic! 
Sorry, Jackie! But not today!
        ''')   
        fail() # failed option 
        
    elif '3' in choice or "BLACKBEARD" in choice  in choice:
        print ("""Usually, he is the one who wants to kill you.
However, he is willing to help you today.""")
        time.sleep(2.4)
        
        print ("""
He will ask you for something later.
But it does not make any sense now! You survived!
        """)
        input("Press Enter to continue...")
        second_move() # right option - we can move to another stage
        
    else:
        print ("""Jackie, please tape the number from 1 to 4 or the name 
(Calipso, Kracken, Blackbeard or Galleon!
        """) ###preventing mistakes in input
        first_move() # wrong input
        
###Function second move introduce some rules for the user in order to pass the next step
 
    
def second_move(): # defining next function
    print ( 80 * '*' )
    print (''' Jackie, by the way in what mood are you today?''')
#just for fun
    mood = input (prompt = ">").upper()
    print ( 80 * '*' ) 
    
    print (f'''Congratulations! 
You are on the board of Blackbeard's boat named '{mood}'! 
He saved your live and needs your help.''')
    time.sleep(2.4)
    
    print ('''
He is asking you to navigate his boat to the island with the Immortality Grail! 
But he needs to know that you are loyal!

He is asking you a question! Don't miss your boat, Jackie!
    ''')
    input ( prompt = "Press Enter to continue...")
    print ( 80 * '*' ) 
   
    print ("""In what year was Blackbeard born? 
I will help you! Try to guess the year. After every attempt 
you will see: Octopuses and Sharks.""")#rules of the next stage
    time.sleep(2.4) 
    
    print ("""
*** Octopus ***  correct number in the correct position
*** Shark *** correct number in the wrong position
Remember, you have 6 tries. """)
    print (80 * '*' ) 
    secret = str(1680)#starting with a secret number = year
    octopus_counter = 0 # starting to count octpuses
    tries = 1 #we need to count tries as we have only 6
    
    while octopus_counter!=4: #starting looping 
        guess = input ("what is your number?")
        if len(guess) == 4 and guess.isdigit() ==True: #only 4 numbers
            shark_counter = 0 
            octopus_counter = 0
            tries += 1
            
            for i in range (len(secret)):#looping the cycle
                if secret[i] == guess[i]: #if the right number is on the right position
                    octopus_counter += 1 #it means that plus one octopus
                elif guess[i] in secret and guess[i]!= secret[i]:#if the right number is on the wrong position
                    shark_counter+=1 #plus one shark
                    
            print( 40 * "_" )
            print ( "you have", octopus_counter,"octopuses", shark_counter, "sharks")
            print ( 40 * "_")
            
            if tries > 6: # in case we have already spent 6 tries - fail the stage
                print('''You have already spent 6 tries! But Blackbeard is super
kind today! He wants to give you the last chance!''')
                second_question()
                
        else:
            print('''There are two options: either you did not put numbers or 
you put less than 4''')   # check for input mistake
            
            
    print(f"""Yes, Jackie! You are fantastic! He was born in {secret} Move on! """) 
    print(80*'*') 
    third_introduction()    # next stage


def second_question():  
    print ( 80 * '*')
    print('''Last chance, Jackie! What was Blackbeard's real name?
I will help you!''')
    
    ed_name = input(prompt= '''Tell me the full name of the movie:  " *** Scissorhands > ''').upper()
    sur_name = input(prompt= "Tell me infinitive of the verb 'teaching'  > ").upper()
     # simple comparison of the name and the surname. No possible option for
    #numbers or other words. User can only enter the right name, otherwise it is fail()
    
    if 'EDWARD' in ed_name  and  'TEACH' in sur_name : #check if words are here
        print ( 80 * '*' ) 
        print ( '''Congratulations! His name is Edward Teach ! Move on!''')
        print ( 80 * '*' ) 
        input (prompt = "Press Enter to continue!")
        third_introduction()
        
    else:
        print("""
        Sorry, Jack Sparrow! You are not loyal gentleman! Bye
        """)
    fail()

    
    
def third_introduction(): # introduction to the next stage
    print ( 80 * '*' )
    print("""You are reaching the island.
But there is a Sirena area.
They want to play a game with you.""")
    time.sleep(2.4)
    
    print ("""
Rules are simple: Sirena and you will roll two dices each. You will have your 
scores and Sirena will have her scores. You need to have a higher total score and 
you need to guess if she tells true about her scores. Good Luck!
 """)
    input ( prompt = "Press Enter to continue!")
    third_stage()
    
    
def third_stage():
    print ( 80 * '*' )
    import random # start with random import
    box1 = 0  #starting counting scores for the user
    box2 = 0  #starting counting scores for Sirena
    tries = 6 # we have only 6 tries
    while tries>0: #starting our loop as we have 6 tries
        cub_1 = (random.randint(1,6)) #creating random numbers from 1 to 6
        cub_2 = (random.randint(1,6))
        cub_3 = (random.randint(1,6))
        cub_4 = (random.randint(1,6))
        res_1 = cub_1 + cub_2 #counting results
        res_2 = cub_3 + cub_4
        print ('Rooooooooling')
        time.sleep(1.2)
        print ("Your rolling scores:", cub_1, cub_2)
        print ( 40 * "_" )
        user = input (prompt= "My number is higher than yours! True? > ").lower()
        if user == "yes" and res_2 > res_1:
            # if Sirena's result is higher than user's and user said 'Yes' - 
            # 0.5 scores to user's box and 0.5 to Sirena's box
            box1 += 0.5
            box2 += 0.5
            tries -= 1 # minus try
            print ( 40 * "_" )
            print ( "Sirena's roll :", cub_3, cub_4)
            print ( 20 * "_ ")
            input (prompt = "Press Enter to continue!")

        elif user == "yes" and res_1 > res_2:
            # if Sirena's result is lower than user's but user said 'Yes' - 
            # 0.5 scores to user's box and 0.5 to Sirena's box
            box1 += 0.5
            box2 += 0.5
            tries -= 1 # minus try
            print ( 40 * "_" )
            print ( "Sirena's rolls:", cub_3, cub_4)
            print ( 20 * "_ " )
            input ( prompt = "Press Enter to continue!")
            

        elif user== "no" and res_1 > res_2:
            # if Sirena's result is lower than user's and user said 'No' - 
            # 1 scores plus to user's box and minus 1 out of Sirena's box
            box1 += 1
            box2 -= 1
            tries -= 1  # minus try
            print ( 40 * "_" )
            print ("Sirena's rolls", cub_3, cub_4)
            print ( 20 * "_ " )
            input ( prompt = "Press Enter to continue!")
            
        elif user== "no" and res_1 < res_2:
            # if Sirena's result is higher than user's and user said 'No' - 
            # no scores to user's box and 1 score to Sirena's box
            box2 += 1
            tries -= 1  # minus try
            print ( 40 * "_" )
            print ( "Sirena's rolls", cub_3, cub_4)
            print ( 20 * "_ " )
            input ( prompt = "Press Enter to continue!")

        elif user == "yes" and res_1 == res_2:
            # if it is tie and user said 'Yes' - 
            # 0.5 scores to user's box and 0.5 to Sirena's box
            box2 += 0.5
            box1 += 0.5
            tries -= 1  
            print ( 40 * "_" )
            print ( "Sirena's rolls", cub_3, cub_4)
            print ( 20 * "_ " )
            input ( prompt = "Press Enter to continue!")

        elif user == "no" and res_1 == res_2:
            # if it is tie and user said 'No' - 
            # 0.5 scores to user's box and 0.5 to Sirena's box
            box1 += 0.5
            box2 += 0.5
            tries -= 1 # minus try
            print ( 40 * "_" )
            print ("Sirena's rolls:", cub_3, cub_4)
            print ( 20 * "_ " )
            input ( prompt= "Press Enter to continue!")

        else:
            print("""I think you did not type "yes" or "no" """) # catch the input mistake
    else: # no more tries, countring results
        print ( "Stop the game!" )
        print ( 20 * "_ ")
        print ( "Your Score", box1, "and", "Sirena's Score", box2)
        print ( 20 * "_ " )

    if box1 > box2: #defining the winner
        print("You won!")
        input( prompt = "Press Enter to continue!")
        final_move()
        
        
    elif box2 > box1: #if Sirena won, we are failed
        print ("Sirena won!")
        fail()
        
    elif box1 ==box2:
        print ("Tie!Once again")
        third_stage()
        
        

import time
def final_move():#simple text function. No logic, just a matter of luck. 
    print ( 80 * '*' )
    print ( """
You have reached the Island with the Grail.
Which path would you like to take?""")
    time.sleep(2.4)
    print ( """
                    
                    ***

1.You can go through forbidden forest and meet your old "friend" - captain Barbossa.
It is the shortest but the most dangerous way!!!""")
    time.sleep(2.4)
    print ( """

                    ***

2.You can go along the seaside.
The longest, however, the safiest way!""")
    time.sleep(2.4)
    print ( """

                    *** 

3.You can take the boat and follow the river.
Probably, you will run into few crocodiles
    """)
    print ( 80 * '*' ) 
    
    final_choice = input (prompt = "Tell me the number ")
    while final_choice != '1': 
    #starting our loop as we have only 3 options
        
        if final_choice == '2':
            print ("""
Unfortunately, the spanish soldiers have already took the Holy 
Grail. I do not think that Blackbeard will like it!
""" )
            fail() 
            
            
        elif final_choice == '3':
            print("""
Or... probably this river was not a right choice.
""")
            fail()
            
        elif final_choice != '1' and final_choice != '2' and final_choice != '3':
            print("""
I think you need to put number 1 or 2 or 3
""") # looking for input mistake
            final_choice = input (prompt = "Tell me the number ")
    
    else:
        print ( """
Good Job, Jackie! Barbossa will try to fight with Blackbeard!
And you are free!
**********************You won!!!******************************************** """)




def fail(): #fail function - continue this game until the user win. However, 
# we can also create fail function with exit of the game (if this requirement is specified)
    print ( 80 * '*' ) 
    print ( "Sorry, Jackie! This is not your day!")
    exit = input ( prompt = ">""Let's try again! Press enter")
    start()
   
```
